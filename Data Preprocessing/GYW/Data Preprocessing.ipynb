```python
import pandas as pd
import os

# 저장할 폴더 지정
folder = 'country_gyw_excel/'
os.makedirs(folder, exist_ok=True)  # 폴더 없으면 생성

# country별로 반복
for country, group in df2.groupby('country'):
    gyw_groups = group['gyw_group'].unique()  # 해당 country의 고유 gyw_group 값
    df_out = pd.DataFrame(gyw_groups, columns=['gyw_group'])  # DataFrame으로 변환
    
    file_path = os.path.join(folder, f'{country}_gyw_group.xlsx')
    df_out.to_excel(file_path, index=False)

    print(f'Saved {file_path}')
```
